```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x = "class" , y = "age", data = df, estimator = "sum")
```

```python
sns.barplot(x = "class", y= "age", hue="sex" , data = df)
```

```python
sns.barplot(x = "class", y= "fare", hue="sex" , data = df, estimator = "sum")
```

```python
sns.violinplot(x = "class", y = "age", data = df)
```

```python
sns.displot(df[df["class"] == "Second"]["age"])
```

# to plot a bar plot

```python
sns.countplot(x = "survived", data = df)
```

```python
sns.countplot(x = "survived", hue = "class" ,data = df)
```

```python
x = pd.crosstab(df["survived"], df["class"])
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, annot = True)
```

```python
sns.heatmap(x, annot = True, fmt = '.1f')
```

```python
sns.heatmap(x, annot = True, fmt = '.1f', cmap = "coolwarm")
```

```python
sns.barplot(x = "class", y= "age", hue="sex" , data = df)
```

```python
sns.stripplot(x = "class", y= "age", hue="sex" , data = df)
```
